```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nombres=["HD19","HD20","HD22"]
trans=[54450.509489,54123.634818,54150.650434]
period=[5.15195,2.4449,1.59495]
```

```python
from astropy.io import ascii

pucheros=ascii.read("pucheros.dat")
coralie=ascii.read("coralie.dat")

```

```python
HD19,HD20,HD22=[],[],[]

def llenar(lista):
    for i in lista:
        if i[0]=="HD198044":
            HD19.append([i[1],i[2]])
        if i[0]=="HD205403":
            HD20.append([i[1],i[2]])
        if i[0]=="HD222891":
            HD22.append([i[1],i[2]])

llenar(pucheros)
llenar(coralie)
```

```python
t19,t20,t22=[],[],[]
r19,r20,r22=[],[],[]

for i in HD19:
    t19.append(i[0])
    r19.append(i[1])
for i in HD20:
    t20.append(i[0])
    r20.append(i[1])
for i in HD22:
    t22.append(i[0])
    r22.append(i[1])
    
```

```python
pi=np.pi
x=np.linspace(2456930.1128699998,2456980.1128699998,1000)
y19=30*np.sin((2*pi/period[0])*x)
%matplotlib inline
plt.plot(t19,r19,".")
plt.plot(x,y19,label="Esperados")
plt.ylim(-40,40)
plt.legend(loc=2)
plt.show()
```

```python
y20=30*np.sin((2*pi/period[1])*x)

plt.plot(t20,r20,".")
plt.plot(x,y20,label="Esperados")
plt.ylim(-40,40)
plt.xlim(5+2.45693e6,45+2.45693e6)
plt.legend(loc=2)
plt.show()
```
